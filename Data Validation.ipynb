```python
import os
import pandas as pd


pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('max_colwidth',100)



kpi=['Labor Force Participation Rate', 'Employment-Population Ratio', 'Unemployment Rate']
files=os.listdir(path)
files.sort()
count_problem=0


for fileName in files:
  file=path+fileName
  df=pd.read_csv(file)
  for i in kpi:
    if i in fileName:
      topic=i
  df['Date']=pd.to_datetime(df['Date']).dt.date
  df = df.reindex(sorted(df.columns), axis=1)
  df.sort_values(['Date'])
  columns=df.columns
  for j in columns:
    if 'website' in j:
      website=j
  df.insert(len(columns),'website-compute',(df[website].str.replace('%','').astype(float)/100) - (df[topic].str.replace('%','').astype(float)/100))
  df_sub=df[df['website-compute']>0]
  df_sub.sort_values(['Date'],inplace=True)
  df_sub.reset_index(drop=True,inplace=True)
  if not df_sub.empty:
    print(f'{len(df_sub)}/{len(df)} differences for {fileName} ({round(len(df_sub)/len(df),4)})')
    display(df_sub)
    print('\n')
    count_problem+=1
  else:
    print(f'all correct for {fileName}')
    print('\n')

print(f'{count_problem}/{len(files)} tables has differences({round(count_problem/len(files),4)})')
```

```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive/')
import warnings
warnings.filterwarnings("ignore")


df0=pd.read_csv('/content/drive/My Drive/BOC/Result/cpsResult_resultQuarterly2003-2021.csv')
df=pd.read_csv('/content/drive/My Drive/BOC/cpsResult_Compare_resultQuarterly2003-2021.csv')
df.drop(columns=['differences'], inplace=True)
df.insert(5, 'int-website', df['int(Median Wage)']-df['website Median Wage'])
df.insert(6, 'sample size', None)
df.insert(7, 'std(wage)', None)
df.insert(8, 'num_filter', None)
for index, row in df.iterrows():
  df0_sub=df0.loc[(df0['year-season']==row['year-season'])&(df0['Age(all means over 16yrs)']==row['Age(all means over 16yrs)'])\
  &(df0['Sex']==row['Sex'])&(df0['Race']==row['Race'])&(df0['Education']==row['Education'])&(df0['Occupation']==row['Occupation'])]
  size=df0_sub['sample size'].item()
  std=df0_sub['std(wage)'].item()
  filter=df0_sub['num_filter'].item()
  df.loc[index, 'sample size']=size
  df.loc[index, 'std(wage)']=std
  df.loc[index, 'num_filter']=filter

tables=list(df['table'].unique())
count_problem=0
info={}
for table in tables:
  df_sub=df[df['table']==table]
  df_sub.sort_values(['year-season'],inplace=True)
  df_sub.reset_index(drop=True,inplace=True)
  df_sub.plot.scatter(x='sample size',y='std(wage)', c='int-website', title=table)

  df_sub_problem=df_sub.loc[abs(df_sub['int-website'])>2]
  df_sub_problem.sort_values(['year-season'],inplace=True)
  df_sub_problem.reset_index(drop=True,inplace=True)

  

  if not df_sub_problem.empty:
    rate=round(len(df_sub_problem)/len(df_sub),3)*100
    name=table+" "+str(len(df_sub_problem))+"/"+str(len(df_sub))
    info[name]=rate
    display(df_sub_problem)
    print('\n')
    count_problem+=1
  else:
    name=table+" "+"0/"+str(len(df_sub))
    info[name]=0


df=df.astype({"sample size": float, "std(wage)": float, "num_filter": float})

df.to_csv('/content/drive/My Drive/BOC/Result/cpsResult_ComparisionResultQuarterly2003-2021_V2.csv', index=False)

print('int-web comparision: abs(diff)>2')
print(f'{count_problem}/{len(tables)} tables has differences: {(round(count_problem/len(tables),3))*100}%')

sorted_info=dict(sorted(info.items(), key=lambda item: item[1], reverse=False))

for k, v in sorted_info.items():
  print(f'abs(int-website)>2 rate for table {k} : {round(v,1)}%' )
```

```python
tables=list(df['table'].unique())
count_problem=0
info=[]
for table in tables:
  df_sub=df[df['table']==table]
  df_sub.insert(5, 'round-website', df_sub['round(Median Wage)']-df_sub['website Median Wage'])
  df_sub_problem=df_sub.loc[abs(df_sub['round-website'])>2]
  df_sub_problem.sort_values(['year-season'],inplace=True)
  df_sub_problem.reset_index(drop=True,inplace=True)
  if not df_sub_problem.empty:
    rate=len(df_sub_problem)/len(df_sub)
    rate='{:.1f}'.format(rate*100)
    info.append('{0}/{1} differences for {2}: {3}%'.format(len(df_sub_problem),len(df_sub), table,rate ))
    display(df_sub_problem)
    print('\n')
    count_problem+=1
  else:
    info.append('0/{0} all correct for {1}'.format(len(df_sub), table))

print('round-web comparision abs(diff)>2')
print(f'{count_problem}/{len(tables)} tables has differences: {(round(count_problem/len(tables),3))*100}%')
for info_ in info:
  print(info_)
```
